```python
import numpy as np
import scipy as sp
import scipy.stats
import matplotlib.pyplot as plt
```

```python
np.seterr(all='raise')
```

```python
# ロジスティックシグモイド関数
sigmoid = np.vectorize(lambda x: 1.0 / (1.0 + np.exp(-x)))
```

```python
# ロジスティック回帰で重みベクトルを求める
# - 引数m0, S0invを与えれば正則化項あり、与えなければ正則化項なしで計算する。
# - 引数w_initで繰り返しの初期値を指定できる。
def logistic_regression(x, t, m0=None, S0inv=None, w_init=None, iter=100):
    Phi = np.column_stack((np.ones(x.shape[0]), x))
    n, dim = Phi.shape
    if m0 is None: m0 = np.zeros(dim)
    if S0inv is None: S0inv = np.zeros((dim, dim))
    if w_init is None: w = np.zeros(dim)
    else: w = w_init.copy()

    history = [(0, w.copy(), 0.0)]
    try:
        for i in range(iter):
            y = sigmoid(w.dot(Phi.T))
            R = (y * (1 - y)).reshape((n, 1))
            grad = Phi.T.dot(y - t) + S0inv.dot(w - m0)
            H = Phi.T.dot(R * Phi) + S0inv
            w -= sp.linalg.inv(H).dot(grad)
            # ↓これを入れないと、条件によってoverflowやunderflowが起こる。
            # 　wの絶対値が大きくならなければ、分母はなんでもよい？
            # w /= np.linalg.norm(w)
            history.append((i + 1, w.copy(), np.linalg.norm(grad)))
    except Exception as e:
        return w, history, e
    return w, history, None
```

```python
# 決定境界のグラフの体裁を整える
def arrange_decision_boundary_plot(ax):
    # ax = plt.gca()
    ax.set_aspect(1)
    ax.set_xlabel('$x_1$')
    ax.set_ylabel('$x_2$')
    ax.set_xlim(-4.0, 8.0)
    ax.set_ylim(-4.0, 6.0)

# データ集合をプロットする
def plot_data(ax, x, t):
    arrange_decision_boundary_plot(ax)
    for i in range(x.shape[0]):
        ax.scatter(x[i, 0], x[i, 1],
                   color = ['red', 'blue'][int(t[i])],
                   marker = ['x', 'o'][int(t[i])])

# 決定境界をプロットする
def plot_decision_boundary(ax, w, **kwargs):
    xs = np.linspace(-4.0, 8.0, 100)
    ax.plot(xs, np.vectorize(lambda x1: - (x1 * w[1] + w[0]) / w[2])(xs), **kwargs)

# 重みベクトルwの履歴のグラフの体裁を整える
def arrange_history_w(ax, iter):
    ax.set_xlabel('$i$')
    ax.set_ylabel('$w$')
    ax.set_xlim(0, iter)

# 重みベクトルwの履歴をプロットする
def plot_history_w(ax, ws, iter, **kwargs):
    arrange_history_w(ax, iter)
    ax.plot([w[0] for w in ws], linestyle='--', **kwargs)
    ax.plot([w[1] for w in ws], linestyle='-.', **kwargs)
    ax.plot([w[2] for w in ws], linestyle=':', **kwargs)

# 誤差関数の勾配の履歴のグラフの体裁を整える
def arrange_history_grad_E(ax, iter):
    ax.set_xlabel('$i$')
    ax.set_ylabel('$log|∇E|$')
    ax.set_xlim(0, iter)

# 誤差関数の勾配の履歴をプロットする
def plot_history_grad_E(ax, grads, iter, **kwargs):
    arrange_history_grad_E(ax, iter)
    ax.plot(np.log10(grads[1:]), **kwargs)
```

```python
# データ集合 その1
# 線形分離可能なケース
N = 50
np.random.seed(0)
t_1 = np.r_[np.ones(N),
            np.zeros(N)]
x_1 = np.r_[sp.stats.multivariate_normal.rvs([2.0, 0.0], [[3.0, 0.8], [0.8, 0.5]], N),
            sp.stats.multivariate_normal.rvs([-1.0, 3.0], [[3.0, 0.8], [0.8, 0.5]], N)]
```

```python
# データ集合 その2
# 線形分離可能でないケース
N = 50
np.random.seed(1)
t_2 = np.r_[np.ones(N),
            np.zeros(N)]
x_2 = np.r_[sp.stats.multivariate_normal.rvs([2.0, 1.0], [[3.0, 0.8], [0.8, 0.5]], N),
            sp.stats.multivariate_normal.rvs([-1.0, 2.0], [[3.0, 0.8], [0.8, 0.5]], N)]
```

```python
plot_data(plt.gca(), x_1, t_1)
```

```python
plot_data(plt.gca(), x_2, t_2)
```

```python
# テスト用パラメータ
ws = [np.array([0.0, 0.0, -1.0]),
      np.array([0.0, 0.5, -1.0]),
      np.array([1.0, 0.5, -1.0]),
      np.array([1.0, 1.0, -1.0])]
m0s = [np.array([0.0, 0.0, -1.0]),
       np.array([0.0, 0.5, -1.0]),
       np.array([1.0, 0.5, -1.0]),
       np.array([1.0, 1.0, -1.0])]
I = np.identity(ws[0].shape[0])

colors = ['r', 'g', 'b', 'y']
```

```python
# テスト用関数
def test(x, t, c, axes, m0=None, S0inv=None, w_init=None, iter=100):
    w, history, e = logistic_regression(x, t, m0=m0, S0inv=S0inv, w_init=w_init, iter=iter)

    plot_decision_boundary(axes[0], w, color=c)
    if m0 is not None: plot_decision_boundary(axes[0], m0, color=c, linestyle='--')
    plot_history_w(axes[1], [h[1] for h in history], iter, color=c)
    plot_history_grad_E(axes[2], [h[2] for h in history], iter, color=c)

    print("w_init = ", w_init, "m0 = ", m0)
    for (i, w_hist, grad) in history[-5:]:
        print(i, w_hist, - w_hist / w_hist[2], grad)
    if e: print(e)
    print()
```

# テスト1

- 正則化項なし。
- wの初期値をいろいろ変えてみる。

```python
# データ集合その1の場合
fig, axes = plt.subplots(1, 3)
plot_data(axes[0], x_1, t_1)
for w_init, c in zip(ws, colors):
    test(x_1, t_1, c, axes, w_init=w_init, iter=100)
```

```python
# データ集合その1の場合 - 繰り返し回数を減らす
fig, axes = plt.subplots(1, 3)
plot_data(axes[0], x_1, t_1)
for w_init, c in zip(ws, colors):
    test(x_1, t_1, c, axes, w_init=w_init, iter=40)
```

```python
# データ集合その2の場合
fig, axes = plt.subplots(1, 3)
plot_data(axes[0], x_2, t_2)
for w_init, c in zip(ws, colors):
    test(x_2, t_2, c, axes, w_init=w_init, iter=100)
```
